```python
data_to_use = q18.drop(['date', 'lights'], axis=1)

data_to_use
```

```python
# creating the independent variable

Xq18 = data_to_use.drop('Appliances', axis=1)

Xq18
```

```python
yq18 = data_to_use['Appliances']

yq18
```

```python
X_train18, X_test18, y_train18, y_test18 = train_test_split(Xq18, yq18, test_size=0.3, random_state=42)

```

```python
scaler = MinMaxScaler()
X_train18 = scaler.fit_transform(X_train18)
X_test18 = scaler.transform(X_test18)
```

```python
model = LinearRegression()
model.fit(X_train18, y_train18)

y_train_pred = model.predict(X_train18)
```

```python
mae18 = mean_absolute_error(y_train18, y_train_pred)


print(f"Mean Absolute Error on the training set: {round(mae18, 3)}")
#this here is just python, usign the fstring,
# learnt that a while ago, became useful here again
```

```python
rmse18 = np.sqrt(mean_squared_error(y_train18, y_train_pred))


print(f"Root Mean Squared Error on the training set: {round(rmse18, 3)}")
```

```python
y_test_pred = model.predict(X_test18)


mae_test18 = mean_absolute_error(y_test18, y_test_pred)


print(f"Mean Absolute Error on the test set: {round(mae_test18, 3)}")
```
